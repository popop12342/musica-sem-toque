```python
def parse_line(l):
    l = l[l.index('x = ')+4:]
    x = float(l[:l.index(',')])
    l = l[l.index('y = ')+4:]
    y = float(l[:l.index(',')])
    l = l[l.index('z = ')+4:]
    z = float(l[:-1])
    return x, y, z
```

```python
time_steps = []
axs = []
ays = []
azs = []
gxs = []
gys = []
gzs = []

with open('signals/sim1.txt', 'r') as f:
    t_step = f.readline()
    while t_step:
        time_steps.append(float(t_step[:-1]))
        acc_line = f.readline()
        ax, ay, az = parse_line(acc_line)
        gyro_line = f.readline()
        gx, gy, gz = parse_line(gyro_line)
        axs.append(ax)
        ays.append(ay)
        azs.append(az)
        gxs.append(gx)
        gys.append(gy)
        gzs.append(gz)
        t_step = f.readline()
        
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(time_steps, axs)
plt.xlabel('Tempo [s]')
plt.ylabel('Accel X [g]')
plt.show()
```

```python
def plot_one_signal(ax, t, s, ylabel):
    ax.plot(t, s)
    ax.set_xlabel('Tempo [s]')
    ax.set_ylabel(ylabel)

fig, axes = plt.subplots(3, 2, figsize=(10, 6))
plot_one_signal(axes[0][0], time_steps, axs, 'Accel X [g]')
plot_one_signal(axes[1][0], time_steps, ays, 'Accel Y [g]')
plot_one_signal(axes[2][0], time_steps, azs, 'Accel Z [g]')
plot_one_signal(axes[0][1], time_steps, gxs, 'Gyro X [dps]')
plot_one_signal(axes[1][1], time_steps, gys, 'Gyro Y [dps]')
plot_one_signal(axes[2][1], time_steps, gzs, 'Gyro Z [dps]')
plt.show()
```
